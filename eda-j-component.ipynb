## Using Forward fill to impute the NUll values

```python
df_1 =df.fillna(method="ffill")
df_1.isnull().sum()
```

```python
df_1.isnull().sum()
```

```python
df_1.head()
```

```python
df_1.describe()
```

## Deleting unnecessary Column

```python
df_1.drop("url", axis = 1, inplace = True)
```

```python
df_1.head()
```

```python
df_1['published_timestamp']=pd.to_datetime(df_1['published_timestamp'])
```

```python
df_1.head()
```

```python
df_1.info()
```

## Visualizations

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_1['is_paid'].value_counts()
```

```python
plt.figure(figsize=(8,6))
plt.title("paid courses and free courses")
sns.countplot(data=df, x= 'is_paid', width= .2)
```

#### Most of the courses are paid

```python
plt.figure(figsize=(8,6))
plt.title("courses and subjects")
sns.countplot(x='subject',data=df_1, width = .4)
```

#### Web Development and Business Finance are the most common courses

## Correlation 

```python
plt.figure(figsize=(8,6))
sns.heatmap(df_1.corr(),annot=True)
```

```python
plt.figure(figsize=(8,6))
plt.title('price vs num_subscribers')
sns.regplot(x='price',y='num_subscribers',color='#C126B2',data=df_1)
```

### We can see their is no relationship between Price and Number of subscriber

```python
plt.figure(figsize=(8,6))
plt.title('price vs content_duration')
sns.regplot(x='price',y='content_duration',color='green',data=df_1)
```

### price in commonly increased by increasing of content duration

###  Relation between content duration and number of subscribers

```python
plt.figure(figsize=(8,6))
plt.title('content_duration vs num_subscribers')
sns.regplot(x='content_duration',y='num_subscribers',color='magenta',data=df_1)
```

### We can see through the visualization there is no relation between number of subscribers and content duration



### Relation between content duration and number of reviews



```python
plt.figure(figsize=(8,6))
plt.title('content_duration vs num_reviews')
sns.regplot(x='content_duration',y='num_reviews',color='orange',data=df)
```

### There is more number of reviews by increasing content duration

## Creating One More feature "Earning" 

```python
df_1['earnings']=df_1['price']*df_1['num_subscribers']
```

```python
df_1.head()
```

```python
plt.figure(figsize=(8,6))
sns.regplot(x='content_duration',y='earnings',color='#C126B2',data=df_1)
```

### There is weak relationship between content Duration and Earnings

## Distripution of earnings in courses subjects

```python
plt.figure(figsize=(10,8))
sns.barplot(x='subject',y='earnings',data=df_1 , width = .4 ,errorbar=('ci',False))
```

### From the Visualizations we can conclude Web Development has the most earnings

### Earnings for every course subject in all levels

```python
plt.figure(figsize=(10,8))
sns.barplot(x='subject',y='earnings',hue='level',data=df_1,dodge=True,errorbar=('ci',False))
```

### We can conclude most earnings is in Web Devolopement especially in All levels content
